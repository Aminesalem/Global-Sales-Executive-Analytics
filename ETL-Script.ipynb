```python
import pandas as pd

# Load the Excel file
file_name = 'Global_Superstore.xlsx'

# Read the 'Orders' sheet
df = pd.read_excel(file_name, sheet_name='Orders')

# 1. Handle missing values in Postal Code
df['Postal Code'] = df['Postal Code'].fillna(0).astype(int).astype(str)

# 2. Fix the dates
df['Order Date'] = pd.to_datetime(df['Order Date'])

# 3. Create a cleaned version
df.to_csv('Global_Sales_Cleaned.csv', index=False)

print("Process finished! Look in the folder icon on the left for 'Global_Sales_Cleaned.csv'")
```
