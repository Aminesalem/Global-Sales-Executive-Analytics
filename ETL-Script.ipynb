# Global Sales ETL Pipeline
This script handles the extraction, transformation, and loading (ETL) of the raw Global Superstore dataset. It ensures data types are optimized and missing values are handled before the data is ingested into Power BI.

```python
import pandas as pd

def run_sales_etl(input_file, output_file):
    print(f"[INFO] Starting ETL pipeline for {input_file}...")
    
    try:
        # 1. EXTRACT
        print("[INFO] Extracting 'Orders' sheet into Pandas DataFrame...")
        df = pd.read_excel(input_file, sheet_name='Orders')
        
        # 2. TRANSFORM
        print("[INFO] Applying transformations and cleaning data...")
        
        # Postal Codes often import as floats due to blank rows. 
        # We fill nulls, convert to integer to drop decimals, then to string for categorical use.
        if 'Postal Code' in df.columns:
            df['Postal Code'] = df['Postal Code'].fillna(0).astype(int).astype(str)
            
        # Ensure robust datetime conversion for Time Intelligence in Power BI
        if 'Order Date' in df.columns:
            df['Order Date'] = pd.to_datetime(df['Order Date'])
            
        # 3. LOAD
        print(f"[INFO] Exporting cleaned dataset to {output_file}...")
        df.to_csv(output_file, index=False)
        
        print("[SUCCESS] ETL pipeline completed. Data is ready for Power BI!")
        
    except FileNotFoundError:
        print(f"[ERROR] The file {input_file} was not found. Please check the directory.")
    except Exception as e:
        print(f"[ERROR] An unexpected error occurred: {e}")

# Execute the pipeline
raw_data_file = 'Global_Superstore.xlsx'
clean_data_file = 'Global_Sales_Cleaned.csv'

run_sales_etl(raw_data_file, clean_data_file)
```
